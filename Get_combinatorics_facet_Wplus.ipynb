```python
#This is just a check: we want every facet (that is not the chosen one or adjacent to it) to be identified with
#0, 1, or 3 other facets.
for h in range(0,120*num_dod):
    X=[]
    for k in Identification_Maps:
        try:
            X.append(k[h])
        except:
            pass
    if len(set(X)) not in [1,3,0] and h%120 not in list(range(12))+[index_chosen_facet]:
        print('ERROR', h, k,len(set(X)))
```

```python
#Here we get a list of which facets become the same one.
Identifications=[]
Removals=[]
for index_dod in range(0,num_dod):
    for i in range(12):
        Removals.append(i)                                                  # vertices that are removed by the gluing
        Removals.append(120+i)      
        for k in range(12):
            Identifications.append([list(Identification_Maps[index_dod*12+i].keys())[k],list(Identification_Maps[index_dod*12+i].values())[k]])  # vertices that are identified by the gluing      
```

```python
#We now find the orbits of these identifications, in order to have a precise list of the facets of the 4-manifold
#with corner angles.
All_vertices=[]
for edge in Identifications:
    All_vertices=All_vertices+edge
All_vertices=[x for x in Set(All_vertices)]
All_vertices.sort()
Identification_Orbits=[  [x] for x in All_vertices]
```

```python
for edge in Identifications:
    for orbit in Identification_Orbits:
        if edge[0] in orbit:
            orbit1=orbit
        if edge[1] in orbit:
            orbit2=orbit
    if orbit1!=orbit2:
        orbit_new=orbit1+orbit2
        Identification_Orbits.append(orbit_new)
        Identification_Orbits.remove(orbit1)
        Identification_Orbits.remove(orbit2)
```

```python
for orbit in Identification_Orbits:
    orbit.sort()
    if len(orbit) not in [2,8,24,num_dod,num_dod*12]:
        if orbit[0]%120 not in list(range(12))+[index_chosen_facet]:
            print("Something is going wrong", len(orbit))
```

```python
#Here we check if there is any facet that share a sub-face with itself. If this condition is not satisfied,
#we cannot colour the manifold with corner in order to obtain a hyperbolic 4-manifold.
#The only dodecahedral L-spaces tessellated with 4 or less hyperbolic right-angled dodecahedra that satisfy
#this check are CubicalOrientableClosedManifold(betti=0)[11] and CubicalOrientableClosedManifold(betti=0)[28].
for edge in S.faces()[1]:
    for x in Identification_Orbits:
        if edge[0] in x and edge[1] in x and edge[0]%120 not in list(range(12))+[index_chosen_facet]:
            print("AUTO-INTERSECTION!")
            print(edge, x)
```

```python
Edges=[]                          
for edge in Graph_edges:
    if edge[0] not in Removals and edge[1] not in Removals:
        Edges.append(edge)
for edge in Edges:
    for orbit in Identification_Orbits:
        if edge[0] in orbit[1:]:
            edge[0]=orbit[0]
        if edge[1] in orbit[1:]:
            edge[1]=orbit[0]
        edge.sort()
```
