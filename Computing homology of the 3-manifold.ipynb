```python
def edge_faces(edge):     # INSERT an edge [a,b] 
    lst=[]
    for i in range(12):
        if edge[0] in Dod[i] and edge[1] in Dod[i]:
            a=Dod[i].index(edge[0])
            b=Dod[i].index(edge[1])
            if abs(a-b)==1 or abs(a-b)==4:
                lst.append([i,[a,b]])
        if len(lst)==2:
            break
    return lst            # RETURNS the faces "x" containing it and the index of vertices in the list Dod[x]
```

```python
def edge_gluing(edge_list):  # INSERT a list of edges [n,[a,b]] and return the copies after applying the gluing maps
    lst=[]                   # this edge connects the vertices "a" and "b" in the dodecahedron "n"
    for edge in edge_list:
        dod=edge[0]
        data=edge_faces(edge[1])
        for edge_face in data:
            for Map in Maps[dod]:
                if edge_face[0] == Map[1][0]:
                    break             # here, Map is gluing the face "edge_face[0]" of the dodecahedron "edge[0]"
            mapping=Map[2][1]       # for the first dodecahedron, the map sends edges through Maps
            lst.append([Map[0][1],[mapping[edge_face[1][0]],mapping[edge_face[1][1]]]])
    return lst   
```

```python
def full_edge_gluing(edge0):        # INSERT an edge and returns the full list of edges identified to that inicial edge
    lst=[edge0]
    while True:
        temp=edge_gluing(lst)
        count=0
        for edge in temp:
            if edge not in lst:
                lst.append(edge)
                count=count+1
        if count==0:                # the cicle breaks when the set is stabilized
            break        
    return lst
```

```python
def vertex_faces(vertex):            # INSERT a vertex p
    lst=[]
    for i in range(12):
        if vertex in Dod[i]:
            lst.append([i,Dod[i].index(vertex)])
        if len(lst)==3:
            break
    return lst                      # RETURNS the faces containing it and the index of vertices in each face
```

```python
def vertex_gluing(vertex_list):  # INSERT a list of vertices [n,p] and return the copies after applying the gluing maps
    lst=[]                       
    for v in vertex_list:
        dod=v[0]
        data=vertex_faces(v[1])
        for vertex in data:
            for Map in Maps[dod]:
                if vertex[0] == Map[1][0]:
                    break             # here, Map is gluing the face "edge_face[0]" of the dodecahedron "edge[0]"
            mapping=Map[2][1]       # for the first dodecahedron, the map sends edges through Maps
            lst.append([Map[0][1],mapping[vertex[1]]])
    return lst   
```

```python
def full_vertex_gluing(vertex0):     # INSERT a vertex and returns the full list of vertices identified to that inicial vertex
    lst=[vertex0]
    while True:
        temp=vertex_gluing(lst)
        count=0
        for vertex in temp:
            if vertex not in lst:
                lst.append(vertex)
                count=count+1
        if count==0:
            break        
    return lst
```

```python
def get_edges_from_face(i):             # INSERT number of face, returns edges of that facet
    lst=[]
    for j in range(4):
        lst.append([Dod[i][j],Dod[i][j+1]])
    lst.append([Dod[i][4],Dod[i][0]])
    return lst
```

```python
def invert_edge(edge):         # inverts the orientation of an edge without changing the list itself
    return list(reversed(edge))
```

```python
def get_edge_rep(face):         # INSERT a face [n,i], the facet of index 'i' on the dodecahedron 'n'
    lst=[0]*num_of_edges
    edges=get_edges_from_face(face[1])
    for j in range(num_of_edges):
        all_edges=Edge_orbits[j]
        for edge in edges:
            if [face[0],edge] in all_edges:
                lst[j]=lst[j]+1
            if [face[0],invert_edge(edge)] in all_edges:
                lst[j]=lst[j]-1
    return lst                  # RETURNS the boundary map on that facet
```
