```python
def get_binary_fixed_length(n,m):
    
    if n>2^(m+1):
        raise Exception("n is too big")
    s=sage.rings.integer.Integer(n).digits(2)
    while len(s)<m:
        s.append(0)
    s.reverse()
    
    return s
```

```python
def get_integer_from_binary_list(bin_list):
    summ=0
    bin_list.reverse()
    for h in range(len(bin_list)):
        summ=summ+2^h*bin_list[h]
    return summ
```

```python
def get_index_from_position_and_colour(position, colour):
    return (get_integer_from_binary_list(colour)*num_120_cell+position)
```

```python
def get_position_and_colour_from_index(index):
    position=mod(index,num_120_cell)
    position=sage.rings.integer.Integer(position)
    colour_integer=int(index-position)//int(num_120_cell)
    col=get_binary_fixed_length(colour_integer,num_col)
    return [position, col]
```

```python
def sum_colours(a,b):
    if len(a)!=len(b):
        raise Exception("Colours have not the same length")
    c=[]
    for h in range(len(a)):
        if a[h]==b[h]:
            c.append(0)
        else:
            c.append(1)
    return c
```

```python
def summ_all(vector, constant):
    res=[]
    for h in vector:
        res.append(h+constant)
    return res
```

```python
#We now move to integer homology. We will create the differential of the chain complexes between degrees 0, 1 and 2.
#These can be used to compute H_1, but Sage could not afford this computation. We use the matrices only to obtain
#the first betti number b_1.
#To obtain the matrices, we want to choose an order of the edges of each pentagon.
```

```python
Pentagonal_orders=[]
x=0
for y in Orbits_faces_2:
    chosen_one=y[2]
    chosen_one_in_first=[mod(a,600) for a in chosen_one]
    edges_pentagon=[]
    for k in s120_faces1:
        if len(set(k) & set(chosen_one_in_first))==2:
            ind0=chosen_one_in_first.index(k[0])
            ind1=chosen_one_in_first.index(k[1])
            edges_pentagon.append([ind0,ind1])
    first_vertex=0
    found=0
    for search_second_vertex in range(1,5):
        if found==0:
            if any([len({first_vertex,search_second_vertex}&set(edg))==2 for edg in edges_pentagon]):
                second_vertex=search_second_vertex
                found=1
        if found==1:
            if any([len({first_vertex,search_second_vertex}&set(edg))==2 for edg in edges_pentagon]):
                if search_second_vertex!=second_vertex:
                    fifth_vertex=search_second_vertex
                    found==2
    found=0
    for search_third_vertex in range(1,5):
        if found==0:
            if any([len({second_vertex,search_third_vertex}&set(edg))==2 for edg in edges_pentagon]):
                if search_third_vertex!=first_vertex:
                    third_vertex=search_third_vertex
                    found=1
    for a in range(0,5):
        if a not in [first_vertex,second_vertex,third_vertex,fifth_vertex]:
            forth_vertex=a
    Pentagonal_orders.append([first_vertex,second_vertex,third_vertex,forth_vertex,fifth_vertex])
    if x%1000==0:
        print(x,len(Orbits_faces_2))
    x=x+1
```

```python
D_C1=C1.dict()
D_C1_keys=D_C1.keys()
Integer_C1=Matrix(C1.nrows(),C1.ncols(),sparse=True)
counterr=0
for roba in D_C1_keys:
    orbit1=Orbits_faces_1[roba[0]]
    orbit2=Orbits_faces_2[roba[1]]
    y=orbit2[0]
    for x in orbit1:
        if len(set(x) & set(y))==2:
            positions=[ y.index(x[0]), y.index(x[1]) ]
            p=Pentagonal_orders[roba[1]]
            if positions==[p[0],p[1]] or positions==[p[1],p[2]] or positions==[p[2],p[3]] or positions==[p[3],p[4]] or positions==[p[4],p[0]]:
                Integer_C1[roba[0],roba[1]]=Integer_C1[roba[0],roba[1]]+1
            elif positions==[p[0],p[4]] or positions==[p[4],p[3]] or positions==[p[3],p[2]] or positions==[p[2],p[1]] or positions==[p[1],p[0]]:
                Integer_C1[roba[0],roba[1]]=Integer_C1[roba[0],roba[1]]-1
            else:
                print(positions,p,roba)
                raise Exception("Something went wrong")
    if counterr%1000==0:
        print(counterr)
    counterr=counterr+1
```

```python
#We now do the same but with the edges (we essentially choose an orientation of each edge)
D_C0=C0.dict()
D_C0_keys=D_C0.keys()
Integer_C0=Matrix(C0.nrows(),C0.ncols(),sparse=True)
counterr=0
for roba in D_C0_keys:
    orbit0=Orbits_faces_0[roba[0]]
    orbit1=Orbits_faces_1[roba[1]]
    for x in orbit0:
        for y in orbit1:
            if len(set(x) & set(y))==1:
                positions=y.index(x[0])
    if positions==0:
        Integer_C0[roba[0],roba[1]]=1
    elif positions==1:
        Integer_C0[roba[0],roba[1]]=-1
    else:
        raise Exception("Something went wrong")
    if counterr%1000==0:
        print(counterr)
    counterr=counterr+1
```
